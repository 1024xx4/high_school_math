```python
# List 8-1 年収 Graph を描画する
import matplotlib.pyplot as plt
import pandas as pd

# salary.csv の読み込み
dat = pd.read_csv('sample_data/salary.csv', encoding='SHIFT-JIS')

# Data を Set
x = dat['年齢']
y = dat['年収']

# Graph を描画
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()

# List 8-2 差分 Graph を描画する
# Data 数
cnt = len(dat)

# 差分をとる
diff_y = []
for i in range(0, cnt-1):
    diff_y.append(y[i+1] - y[i])

# Graph を描画
plt.plot(x[1:], diff_y)
plt.grid(color='0.8')
plt.show()
```
