```python
# List 8-1 年収 Graph を描画する
import matplotlib.pyplot as plt
import pandas as pd

# salary.csv の読み込み
dat = pd.read_csv('sample_data/salary.csv', encoding='SHIFT-JIS')

# Data を Set
x = dat['年齢']
y = dat['年収']

# Graph を描画
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()

# List 8-2 差分 Graph を描画する
# Data 数
cnt = len(dat)

# 差分をとる
diff_y = []
for i in range(0, cnt - 1):
    diff_y.append(y[i + 1] - y[i])

# Graph を描画
plt.plot(x[1:], diff_y)
plt.grid(color='0.8')
plt.show()
```

```python
# List 8-3 関数と導関数の Graph を描画する
import matplotlib.pyplot as plt
import numpy as np

# x の値
x = np.arange(-10, 10, 0.1)

# 元の関数
y = x**3 + 3*x**2 + 3*x +1

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()

# 導関数
y2 = 3*x**2 + 6*x + 3
plt.plot(x, y2)
plt.grid(color='0.8')
plt.show()
```
