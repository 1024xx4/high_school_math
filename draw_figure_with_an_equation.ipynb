```python
# List 3-4 Solve simultaneous equations.
from sympy import Symbol, solve

# Define expression.
a = Symbol('a')
b = Symbol('b')
ex1 = a + b - 1
ex2 = 5 * a + b - 3

# Solving simultaneous equations.
print(solve((ex1, ex2)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Slope and intercept of the underlying line segment.
# Vertical bisector of the line segment connecting (0,1)to(6,5)
a1 = (5 - 1) / (6 - 0)
b1 = 1

# Midpoint of line segment.
cx = (0 + 6) / 2
cy = (1 + 5) / 2

# Slope of a straight line orthogonal to a line segment.
a2 = -1 / a1

# Intercept of a straight line orthogonal to a line segment.
b2 = cy - a2 * cx

# Equation of a straight line.
x = np.arange(0, 7)  # Value of a x
y1 = a1 * x + b1  # Original line segment.
y2 = a2 * x + b2  # Vertical bisector.

# Drawing.
plt.plot(x, y1)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# List 3-7 Draw circle using trigonometric ratio.
import matplotlib.pyplot as plt
import numpy as np

# Angle.
th = np.arange(0, 360)

# Coordinates of point p on the circumference.
x = np.cos(np.radians(th))
y = np.sin(np.radians(th))

# Drawing.
plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()

# Draw a circle around the coordinates (2,3).
x = np.cos(np.radians(th)) + 2
y = np.sin(np.radians(th)) + 3

plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()

# Draw a circle of the radius other than 1.
r = 5
x = r * np.cos(np.radians(th))
y = r * np.sin(np.radians(th))

plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# List 3-8 Draw of a circle with a radius of 300(y >= 0)
import matplotlib.pyplot as plt
import numpy as np

# Circle equations.
r = 300  # Radius.
x = np.arange(-r, r * 1)  # x: -300 ～ 300
y = np.sqrt(r ** 2 - x ** 2)

# Draw.
plt.plot(x, y)
plt.axis('equal')
plt.show()
```

```python
# Draw a circle with the center(200, 300).
import matplotlib.pyplot as plt
import numpy as np

# Center of circle.
a = 200
b = 300

# Circle equations
r = 300  # Radius
x = np.arange(a - r, a * r * 1)  # x: Range required to draw a circle with a radius of 300 centered on (200, 300)
y = np.sqrt(r ** 2 - (x - a) ** 2) + b  # y: Above the circle
y2 = -y + 2 * b  # y2: Below the circle

# Draw
plt.plot(x, y)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
# Heron's formula
import math

x = [1, 3, 6]  # x coordinate(A, B, C in that order)
y = [5, 1, 4]  # y coordinate(A, B, C in that order)
a = math.sqrt((x[1] - x[0]) ** 2 + (y[1] - y[0]) ** 2)  # Length of side AB
b = math.sqrt((x[2] - x[1]) ** 2 + (y[2] - y[1]) ** 2)  # Length of side BC
c = math.sqrt((x[2] - x[0]) ** 2 + (y[2] - y[0]) ** 2)  # Length of side AC
s = (a + b + c) / 2
math.sqrt(s * (s - a) * (s - b) * (s - c)) # Heron's formula
```
